# Sistemas de Terceira Ordem
Dois polos complexos conjugados e um polo real.


```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
#Pode ser utilizado "Python Control Systems Library"
```

```python
#Parametros so sistema (de acordo com a bibliografia)
Wn = 1;
ep = 0.7;
beta = [0.5, 1, 2, 4, 100];
```

## função de transferencia

$G=\dfrac{\omega_n^2p}{(s^2+2\zeta\omega_ns+\omega_n^2)(s+p)}$

```python
tm = np.linspace(0,20,100) #criar um temporizador de 0 até 20 (Time (s))
for i in beta:
    p = i*ep*Wn;
    num = [Wn**2*p]; #parametros numerador da funcao de transferencia
    den = np.convolve([1, 2*ep*Wn, Wn**2], [1, p]) #parametros denominador da funcao de transferencia
    G = signal.TransferFunction(num,den) #função de tranferência (procedimento)
    t,y = signal.step(G,None,tm,20) #sinal do degrau unitário aplicado sobre o sistema
    plt.figure(1) # diretivas de plotagem
    plt.plot(t,y,label=r'$\beta=$'+str(i))
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.title('Step response')
plt.grid()
plt.legend()
```
