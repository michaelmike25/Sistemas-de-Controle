```python
import numpy as np
import matplotlib.pyplot as plt

def y(t):
    return 1.0 - 3*np.exp(-t) + 3*np.exp(-2*t) - np.exp(-3*t)

t = np.linspace(0, 8, 400)
yt = y(t)

plt.figure(figsize=(8,4.5))
plt.plot(t, yt, linewidth=2)
plt.axhline(1, linestyle='--', linewidth=1, label='y(∞)=1')
plt.scatter([0,0.5,1,2,3,5],[0,0.061,0.253,0.646,0.858,0.980], color='k')
plt.vlines([1,2,3], ymin=0, ymax=1.05, linestyles=':', label='polos: τ=1,0.5,0.333')
plt.ylim(-0.05,1.05)
plt.xlim(0,8)
plt.xlabel('t (s)')
plt.ylabel('y(t)')
plt.title('Resposta ao degrau da função no tempo: y(t)=1-3e^{-t}+3e^{-2t}-e^{-3t}')
plt.grid(True)
plt.legend()
plt.show()
```
